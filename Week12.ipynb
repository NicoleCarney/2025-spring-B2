# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

The side-by-side subplots clearly separate the two weeks, which helps if I want to focus on each week individually. But if the goal is to directly compare patterns day by day, the single chart with two lines is more effective. It lets you instantly see where one week outperformed the other, especially if the lines are color coded and labeled clearly.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for survey results (Survey items x Response categories)
survey_items = ['Survey item A', 'Survey item B', 'Survey item C', 'Survey item D']
data = np.array([
    [1, 5, 10, 33, 51],  # Survey item A
    [5, 12, 20, 30, 33],  # Survey item B
    [8, 9, 18, 27, 38],   # Survey item C
    [11, 9, 20, 28, 32]   # Survey item D
])

# Categories
labels = ['Strongly Disagree', 'Disagree', 'Neutral', 'Agree', 'Strongly Agree']
colors = ['#f44336', '#e57373', '#9e9e9e', '#4caf50', '#009688']

fig, ax = plt.subplots(figsize=(10, 6))
bottom = np.zeros(len(survey_items))

# Stack the bars
for i in range(len(labels)):
    ax.bar(survey_items, data[:, i], bottom=bottom, color=colors[i], label=labels[i])
    bottom += data[:, i] 

ax.set_ylabel('Percent of Total')
ax.set_title('Survey Results: Team X')
ax.legend(title='Survey Response')
plt.tight_layout()
plt.show()
```
